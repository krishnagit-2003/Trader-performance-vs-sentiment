```python
#Convert 'date' in sentiment to datetime in fear_greed dataframe
fear_greed['date'] = pd.to_datetime(fear_greed['date'])
```

```python
# Convert 'Timestamp IST' in historical data to datetime
historical['Timestamp IST'] = pd.to_datetime(historical['Timestamp IST'], format='%d-%m-%Y %H:%M')
```

```python
# Extract just the date for mapping
historical['date'] = historical['Timestamp IST'].dt.date
fear_greed['date'] = fear_greed['date'].dt.date
```

```python
# Merge datasets on date
merged = pd.merge(historical, fear_greed[['date', 'classification']], on='date', how='left')
```

```python
# Drop rows without classification
merged = merged.dropna(subset=['classification'])
```

**Feature Engineering**

```python
# Define winning trade (PnL > 0)
merged['Win'] = merged['Closed PnL'] > 0
```

```python
# Group by sentiment classification
performance_summary = merged.groupby('classification').agg({
    'Closed PnL': ['mean', 'sum'],
    'Win': ['mean', 'count'],
    'Fee': 'mean',
    'Size USD': 'mean'
}).reset_index()
```

```python
performance_summary.columns = ['Sentiment', 'Avg_PnL', 'Total_PnL', 'Win_Rate', 'Num_Trades', 'Avg_Fee', 'Avg_Trade_Size']
```

```python
performance_summary.shape
```

 **Visualizations**

```python
sns.set(style="whitegrid")
```

```python
# Boxplot of PnL by sentiment
plt.figure(figsize=(10, 6))
sns.boxplot(x='classification', y='Closed PnL', data=merged)
plt.title('Trader PnL Distribution by Market Sentiment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Barplot of average PnL by sentiment
plt.figure(figsize=(8, 5))
sns.barplot(x='Sentiment', y='Avg_PnL', data=performance_summary)
plt.title('Average Trader PnL by Sentiment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Barplot of win rate by sentiment
plt.figure(figsize=(8, 5))
sns.barplot(x='Sentiment', y='Win_Rate', data=performance_summary)
plt.title('Win Rate by Market Sentiment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram of leverage (if applicable)
if 'Leverage' in merged.columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=merged, x='Leverage', hue='classification', element='step', stat='density', common_norm=False)
    plt.title('Leverage Distribution by Sentiment')
    plt.tight_layout()
    plt.show()
else:
    print("'Leverage' column not found in the dataset. Skipping leverage plot.")
```

```python
# --- Export insights ---
performance_summary.to_csv("sentiment_vs_performance_summary.csv", index=False)
display(performance_summary.head())
print("\nSummary exported to 'sentiment_vs_performance_summary.csv'")
```
